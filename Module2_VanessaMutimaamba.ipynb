```python
# Lets count the occurences of different weather and road conditions
print(df['Weather_conditions'].value_counts())
print(df['Road_conditions'].value_counts())
print(df['Day'].value_counts())
print(df['Light_conditions'].value_counts())
```

```python
#Grouping accidents by speed limits to see where the highest average casualties occur
speed_stats=df.groupby('Speed_limit')[['Number_of_Casualties','Number_of_Vehicles']].mean()
speed_stats
print(speed_stats.loc[[20,30,40,50,60,70]])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Set size and style of the chart
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
#Create a barchart counting accidents by weather condition
#We use top 5 weather conditions so its not too cluttered
top_weather=df['Weather_conditions'].value_counts().head(5)
sns.barplot(x=top_weather.index,y=top_weather.values,palette='viridis')
#Add titles and labels
plt.title('Top 5 weather conditions during accidents',fontsize=16)
plt.xlabel('Weather Condition',fontsize=12)
plt.ylabel('Number of Accidents',fontsize=12)
#Show the plot
plt.show()
```
